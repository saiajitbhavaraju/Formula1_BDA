```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import s3fs

# --- 1. Configure S3 (MinIO) Connection ---
endpoint_url = 'http://minio:9000'
s3 = s3fs.S3FileSystem(
    client_kwargs={'endpoint_url': endpoint_url},
    key='minioadmin',
    secret='minioadmin'
)

# --- 2. Read Processed Data ---
file_path = 's3://processed-data/bahrain_stint_analysis.parquet'
with s3.open(file_path) as f:
    df = pd.read_parquet(f)

print(df.head())

# --- 3. Plot ---
plt.figure(figsize=(15, 8))
sns.barplot(
    data=df.query("LapsInStint > 10"), 
    x="Driver", 
    y="AvgLapTime", 
    hue="Stint"
)
plt.title("Average Lap Time per Stint (Min 10 Laps)")
plt.ylabel("Average Lap Time (s)")
plt.xticks(rotation=45)
plt.show()
```
